```

X_train,X_test,Y_train,Y_test = train_test_split(dfx,dfy,test_size=0.2)
```

```

def hypothesis(X,theta,Y):
  return np.dot(X,theta)-Y
```

```
def error(X,theta,Y):
  m = X.shape[0]
  h0 = hypothesis(X,theta,Y)
  error = ( np.sum(np.square(h0)) )/(2*m)
  return error
```

```

def grad(theta,X,Y):
  h0 = hypothesis(X,theta,Y)
  return np.dot(X.T,h0)
```

```
def grad_des(theta,X,Y,learning_rate=0.1):
 # print(theta,X,Y)
  m=X.shape[0]
  err_list=[]
  for itr in range(10000) :
    theta = theta - learning_rate*grad(theta,X,Y)/m
    err = error(X,theta,Y)
    print(err)
    err_list.append(err)
  return theta,err_list
```

```
theta = np.random.randn(6,1)
print(theta)
```

```
#theta = np.zeros((6,1))
print(theta.shape)
theta,err_list = grad_des(theta,X_train,Y_train)
```

```
Y_pred = hypothesis(X_train,theta,Y_train) + Y_train
```

```
def coeff_of_det(YA,YP):
  temp = (YA-YP)*(YA-YP)/((YA-YA.mean(axis=0))*(YA-YA.mean(axis=0)))
  return 1-temp
```

```
R = coeff_of_det(Y_train,Y_pred)
print(R,R.shape)
```
